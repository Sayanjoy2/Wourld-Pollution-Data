Import Pandas

```python
import pandas as pd
```

Import Data

```python
df = pd.read_csv('./Woruld_data.csv')
df.head(4)
```

Find null values in Data

```python
df.isnull().sum()
```

Drop find null values

```python
df.dropna(inplace= True)
```

```python
df.isnull().sum()
```

```python
df.head(4)
```

Drop and Arrange column

```python
df.drop('country.1',axis=1)
```

```python
df = df[['continent','country','capital_city','mega_city','capital_population','population','total_co2','co2_per_capita','total_methane','methane_per_capita']]
df.head(2)
```

```python
df['total_co2'] = df['total_co2'].str.replace("bn t","")
df.head()
```

```python
df['total_co2'] = df['total_co2'].str.replace("m t","")
df.head(10)
```

Convert data to numeric

```python
df['total_co2'] = pd.to_numeric(df['total_co2'])
df.dtypes
```

```python
df['co2_per_capita'] = df['co2_per_capita'].str.replace("t", "")
df['co2_per_capita'] = pd.to_numeric(df['co2_per_capita'])
```

```python
df.head(4)
```

```python
df['total_methane'] = df['total_methane'].str.replace('m t','')
```

```python
df['total_methane'] = df['total_methane'].str.replace('bn t','')
```

```python
df['total_methane'] = pd.to_numeric(df['total_methane'])
```

```python
df['methane_per_capita'] = df['methane_per_capita'].str.replace('t','')
```

```python
df['methane_per_capita'] = pd.to_numeric(df['methane_per_capita'])
```

```python
df.dtypes
```

```python
df = df.to_csv('World_Data0.csv')
```
